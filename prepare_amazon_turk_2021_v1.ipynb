```python
len(g1_idx), len(g2_idx)
```

```python
g1_terms = np.array(terms)[g1_idx]
np.random.shuffle(g1_terms)
g2_terms = np.array(terms)[g2_idx]
np.random.shuffle(g2_terms)

g3_terms = np.flip(g1_terms)
g4_terms = np.flip(g2_terms)
```

```python
print(np.where(g1_terms == g2_terms))
print(np.where(g1_terms == g3_terms))
print(np.where(g1_terms == g4_terms))

print(np.where(g2_terms == g3_terms))
print(np.where(g2_terms == g4_terms))

print(np.where(g3_terms == g4_terms))
```

```python
data = {
    "PROMPT01" : list(g1_terms),
    "PROMPT02" : list(g2_terms),
    "PROMPT03" : list(g3_terms),
    "PROMPT04" : list(g4_terms),
}
```

```python
df = pd.DataFrame.from_dict(data)
df
```
