```python
reload(rd)
import base64
import json

def get_web_data(idx_pair, part_idx=5, part_name="angel bodies", part_idx_to_show=[]):    
    supporting = []
    for idx in idx_pair:

        drawing_raw = obj['train_data'][idx]
        absolute_coord = rd.to_absolute(drawing_raw)
        stroke5 = rd.absolutedata2stroke5(absolute_coord, label_selected=[part_idx])
        if len(stroke5) < 5:
            print(idx)
#             continue
        
#         print("\n")
#         fig = plt.figure(figsize=(5,5))
#         image_data = rd.transform_spg_2_svg_png(drawing_raw, draw_color=True, stroke_width = 3)
#         plt.imshow(image_data)
#         plt.show()
        
        
#         image_data_other = rd.transform_5_stroke_2_svg_png(stroke5)
#         plt.figure(figsize=(5,5))
#         plt.imshow(image_data_other)
        
        drawing_dict = {}
        
        if len(part_idx_to_show) < 1:
            part_idx_to_show = np.unique(np.asarray(absolute_coord)[:,-1])
        
        for l in part_idx_to_show:
            stroke3_selected = rd.select_absolutedata(absolute_coord, label_selected=[l])
            stroke3_selected_split = rd.split_into_individual_strokes(stroke3_selected)
            drawing_dict[int(l)] = stroke3_selected_split
        supporting.append(drawing_dict)


    data_dict = {
        "Qtype" : "img-sketch",
        "Question" : "angel",
        "Supporting" : supporting,
        "PartsToAnnotate" : [[part_idx, part_name]],
    }
    return data_dict
```

```python
num_sketches_per_annotation = 20
num_annotations = total_num // num_sketches_per_annotation 
if num_annotations * num_sketches_per_annotation < total_num:
    num_annotations += 1
wing_list = []
for anno_idx in range(5):
    start_idx = anno_idx * 20
    end_idx = (anno_idx+1) * 20
    if end_idx > total_num:
        end_idx = total_num
    wing_list.append(list(range(start_idx, end_idx)))

    
data = []
data_raw = []
for l in wing_list:
    l_dict = get_web_data(l, part_idx=7, part_name="angel wings", part_idx_to_show=[7])
    data.append(l_dict)
    
y = json.dumps(data)
dataBytes = y.encode("utf-8")
encoded = base64.b64encode(dataBytes)
output_string = str(encoded)[2:-1]

amazon_data = {
    "base64" : [output_string],
}
df = pd.DataFrame.from_dict(amazon_data)
df.to_csv('/home/xiaoyuz1/2022_0202_pilot_dummy.csv', index=False)  
# df.to_csv('/raid/xiaoyuz1/amazon_turk/2022_0202_pilot_dummy/round1.csv', index=False)  
output_string
```

```python
def draw_pics(idxs):
    category_name = "angel"
    obj = json.load(open(
    '/raid/xiaoyuz1/sketch_datasets/SketchX-PRIS-Dataset/Perceptual Grouping/{}.ndjson'.format(category_name), 
        'r'))
    fig = plt.figure(figsize=(10,10))
    index = 0
    for idx in idxs:
        plt.subplot(2,2,index+1);
        drawing_raw = obj['train_data'][idx]
        absolute_coord = rd.to_absolute(drawing_raw)
        stroke5 = rd.absolutedata2stroke5(absolute_coord, label_selected=[5])
        
        image_data = rd.transform_spg_2_svg_png(drawing_raw, draw_color=True, stroke_width = 3)
        plt.imshow(image_data)

        index += 1
    plt.show()
```
