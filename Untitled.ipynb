```python
encoder_out = model.encoder(img_b)
batch_size = encoder_out.size(0)
encoder_out = encoder_out.view(batch_size, -1, model.enc_hidden_size) #(B, H*W, C)
mean_encoder_out = encoder_out.mean(dim=1) #(B, C)
y = model.fc_params(mean_encoder_out) #(B, 6*M)

params = torch.split(y, 6, 1)
params_mixture = torch.stack(params) #(M, B, 6)

pi,mu_x,mu_y,sigma_x,sigma_y,rho_xy = torch.split(params_mixture,1,2) #(M, B, 1)
pi = F.softmax(pi.transpose(0,1).squeeze(2), dim=1).view(-1, model.M) # (B, M)
sigma_x = torch.exp(sigma_x.transpose(0,1).squeeze()).view(-1, model.M)
sigma_y = torch.exp(sigma_y.transpose(0,1).squeeze()).view(-1, model.M)
rho_xy = torch.tanh(rho_xy.transpose(0,1).squeeze()).view(-1, model.M)
mu_x = mu_x.transpose(0,1).squeeze().contiguous().view(-1, model.M)
mu_y = mu_y.transpose(0,1).squeeze().contiguous().view(-1, model.M)
```

```python
pi = pi.detach().numpy().reshape(-1,)
sigma_x = sigma_x.detach().numpy().reshape(-1,)
sigma_y = sigma_y.detach().numpy().reshape(-1,)
rho_xy = rho_xy.detach().numpy().reshape(-1,)
mu_x = mu_x.detach().numpy().reshape(-1,)
mu_y = mu_y.detach().numpy().reshape(-1,)
```

```python
def sample_bivariate_normal(mu_x,mu_y,sigma_x,sigma_y,rho_xy, greedy=False):
    mean = [mu_x, mu_y]
    #sigma_x *= np.sqrt(0.25)
    #sigma_y *= np.sqrt(0.25)
    cov = [[sigma_x * sigma_x, rho_xy * sigma_x * sigma_y],\
        [rho_xy * sigma_x * sigma_y, sigma_y * sigma_y]]
    x = np.random.multivariate_normal(mean, cov, 1)
    return x[0][0], x[0][1]
```

```python
all_samples = []
for i in range(1000):
    # pi_idx = np.argmax(pi) #np.random.choice(hp.M, p=pi)
    
    acc_pis = [np.sum(pi[:i]) for i in range(1, len(pi)+1)]
    # assert np.isclose(acc_pis[-1], 1)
    # sample uniform
    r = np.random.uniform(0, 1)
    # select Gaussian
    k = 0
    for i, threshold in enumerate(acc_pis):
        if r < threshold:
            k = i
            break
    
    
    x,y = sample_bivariate_normal(mu_x[k],mu_y[k],sigma_x[k],sigma_y[k],rho_xy[k])
    all_samples.append([x,y])
```
